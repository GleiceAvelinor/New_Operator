# 🎥 Reconhecimento de Gestos em Tempo Real

Este notebook utiliza o modelo treinado (`gesture_model.pkl`) e o MediaPipe para identificar gestos capturados pela webcam em tempo real.

```python
import cv2
import mediapipe as mp
from mediapipe.tasks import python as mp_python
from mediapipe.tasks.python import vision
import numpy as np
import joblib
import os

print("✅ Bibliotecas carregadas!")
```

## ⚙️ Configurações e Carregamento do Modelo

```python
CONFIG = {
    "model_pkl": "gesture_model.pkl",
    "mp_task": "gesture_recognizer.task",
    "num_hands": 1
}

if not os.path.exists(CONFIG["model_pkl"]):
    print(f"❌ Erro: O modelo treinado '{CONFIG['model_pkl']}' não foi encontrado!")
    print("💡 Execute o notebook 'gesture_training.ipynb' primeiro.")
else:
    # Carregar o modelo de Machine Learning
    clf = joblib.load(CONFIG["model_pkl"])
    print(f"📦 Modelo ML carregado com sucesso!")

    # Carregar o Recognizer do MediaPipe (apenas para extração de marcos)
    base_options = mp_python.BaseOptions(model_asset_path=CONFIG["mp_task"])
    options = vision.GestureRecognizerOptions(base_options=base_options, num_hands=CONFIG["num_hands"])
    recognizer = vision.GestureRecognizer.create_from_options(options)
    print("✅ MediaPipe inicializado!")
```

```python
def draw_prediction(img, label, confidence):
    """Exibe o resultado da predição na tela."""
    text = f"{label} ({confidence:.2%})"
    cv2.rectangle(img, (10, 10), (450, 70), (0, 0, 0), -1)
    cv2.putText(img, text, (20, 50), cv2.FONT_HERSHEY_DUPLEX, 1.0, (0, 255, 0), 2)

def extract_landmarks(hand_landmarks):
    """Converte os landmarks do MediaPipe para o formato que o modelo espera."""
    return [v for lm in hand_landmarks for v in (lm.x, lm.y, lm.z)]
```
